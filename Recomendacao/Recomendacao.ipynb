```python
# Criando as regras de associação
regras = association_rules(frequent_productsets, metric="lift")
```

## **Modelo Recomendação**

```python
from scipy.spatial.distance import cdist
```

```python
product_frequency = dataset['produto'].value_counts().reset_index()
product_frequency.columns = ['produto', 'frequencia_compra']
product_name_mapping = {idx: product for idx, product in enumerate(product_frequency['produto'])}
product_name_mapping
```

```python
dataset = dataset.merge(product_frequency, on='produto', how='left')
```

```python
def recommend_products(num_recommendations=5):

    dataset_clients_recomend = pd.DataFrame(columns=['email_participante', 'produtos_recomendados']) # o dataset final que será retornado no final
    for client_id in dataset['email_participante']:
        client_cluster = client_agru[client_id]
        cluster_products = np.where(client_agru == client_cluster)[0]
        
        # Filtrar as regras de associação relevantes para o cluster
        relevant_regras = regras[regras['antecedents'].apply(lambda x: any(product in cluster_products for product in x))]
    
        # Calcular as distâncias do usuário para os centróides dos clusters
        client_distances = cdist(dados_agru_two_dim[client_id].reshape(1, -1), centroides, metric='cosine')[0]
        
        # Ordenar os produtos pelo grau de similaridade com os centróides
        recommended_indices = cluster_products[np.argsort(client_distances)]
        
        # Selecionar os produtos recomendados, excluindo os já comprados
        recommended_products = [product_name_mapping[product_idx] for product_idx in recommended_indices if transactions.iloc[client_id][product_idx] == 0]
        
        # Filtrar produtos recomendados com base nas regras de associação
        for rule in relevant_regras.iterrows():
            antecedents = rule[1]['antecedents']
            consequents = rule[1]['consequents']
            
            for antecedent in antecedents:
                for consequent in consequents:
                    if antecedent in recommended_products:
                        recommended_products.remove(antecedent)
                    if consequent in recommended_products:
                        recommended_products.remove(consequent)

        inv_mapping = {v: k for k, v in le_mapping['email_participante'].items()}
        client_email = inv_mapping[client_id]
        # Adicionar cada produto recomendado como uma linha separada no DataFrame
        for recommended_product in recommended_products[:num_recommendations]:
                inv_mapping = {v: k for k, v in le_mapping['produto'].items()}
                product_name  = inv_mapping[recommended_product]
            dataset_clients_recomend = dataset_clients_recomend.append({'email_participante': client_email, 'produto_recomendado': product_name }, ignore_index=True)
        return dataset_clients_recomend


```

```python
recommend_products = recommend_products()
print("Itens recomendados:", recommend_products)
```
